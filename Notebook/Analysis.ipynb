```python
# Histograma da distribuio de preos

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True)
plt.xlabel('Preo')
plt.ylabel('Contagem')
plt.title('Histograma da Distribuio de Preos dos Veculos')
plt.show()
```

```python
# Grfico de disperso: Preo vs. Ano do Modelo

data = pd.read_csv("vehicles.csv")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='model_year', y='price', hue='condition')
plt.xlabel('Ano do Modelo')
plt.ylabel('Preo')
plt.title('Grfico de Disperso: Preo vs. Ano do Modelo')
plt.legend(title='Condio')
plt.show()
```
