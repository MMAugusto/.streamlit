```python
# Histograma da distribuição de preços

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True)
plt.xlabel('Preço')
plt.ylabel('Contagem')
plt.title('Histograma da Distribuição de Preços dos Veículos')
plt.show()
```

```python
# Gráfico de dispersão: Preço vs. Ano do Modelo

data = pd.read_csv("vehicles.csv")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='model_year', y='price', hue='condition')
plt.xlabel('Ano do Modelo')
plt.ylabel('Preço')
plt.title('Gráfico de Dispersão: Preço vs. Ano do Modelo')
plt.legend(title='Condição')
plt.show()
```
